## **Step2: Fetch The Data**
**Download the Canadian Immigration dataset and read it into a pandas dataframe.**


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"
data = pd.read_csv(url)
data.head()
```

Make the indicies of the data frame the **country** column

```python
data.set_index('Country',inplace=True)
data.head()
```

## **Step3: Waffle charts  using pyWaffle package**

```python
#install pywaffle
!pip install pywaffle
```

```python
from pywaffle import Waffle
```

## We will display the proportiona of Denmark, Norway, and Sweden total immigrant contribution.

```python
# lets do a new dataframe for those countries
df_dns = data.loc[['Denmark','Norway','Sweden'],:]
df_dns
```

```python
#lets make a waffle charts
fig = plt.figure(FigureClass = Waffle,rows=20,columns=30,values = df_dns['Total'],cmap_name='tab20',
                 legend={'labels':[f"{k} ({v})" for k,v in zip(df_dns.index.values,df_dns.Total)],'loc':'lower left','bbox_to_anchor':(0,-0.1),'ncol':3})

#show the chart
plt.show()

```

Lets display the proportiona of China and Inida total immigrant contribution.

```python
# Make the data frame for this two country
df_CI = data.loc[['China','India'],:]
# df_CI


#waffle charts
fig2 = plt.figure(FigureClass = Waffle,rows=20,columns=30,values=df_CI['Total'],cmap_name='tab20',
                  legend={'labels':[f"{k} ({v})" for k,v in zip(df_CI.index.values,df_CI.Total)],
                          'loc':'lower left',
                          'ncol':2,
                          'bbox_to_anchor':(0,-0.1)})
plt.show()
```
